```r
triangle <- function(a,delta){
    y <- exp(-100 * (a-0.5)^2)*delta
    return(y)
}

mu.mod <- function(a,l,delta){
    mu <- as.numeric(l%*%c(0.2,0.2,0.3,-0.1))+triangle(a, delta)+a*(-0.1*l[,1]+0.1*l[,3]) + 1
    return(mu)
}

pifunc <- function(a,l){
       l <- as.matrix(l)
       mu <- as.numeric(l%*%c(0.1,0.1,-0.1,0.2)) + 0.5
       sd = 1.5
       p = dnorm(a, mu, sd) / (pnorm(1, mu, sd) - pnorm(0, mu, sd))
       return(p)
}
```
